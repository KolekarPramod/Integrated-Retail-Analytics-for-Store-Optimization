## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.
features_df['Date'] = pd.to_datetime(features_df['Date'],dayfirst=True)
sales_df['Date'] = pd.to_datetime(sales_df['Date'],dayfirst=True)
````

````python
features_df.info()
````

````python
# Merge different datasets
df = pd.merge(sales_df, features_df, how = 'left', on = ['Store','Date','IsHoliday']) # merging sales and features by left since there are more rows for sales
df_new = pd.merge(df, stores_df, how = 'left', on = 'Store') #merge df and df_stores by left to keep all the columns and rows
df_new = df_new.fillna(0)
````
